# Verify output of Prophet webapp

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
filename = './prophet_output/test_result.csv'
df = pd.read_csv(filename)
print(df.head())
print(df.tail())
```

```python
df.plot(x='ds', xlabel='Time', ylabel='Flows', title='Forecast', rot=45)
```

```python
# Dataset 2 years + 1 year forecast
pos = 365*2 

ax = df.plot(x='ds', y=['yhat', 'trend', 'trend_lower', 'trend_upper'], xlabel='Time', ylabel='Flows', title='Forecast', rot=45)
ax.axvline(x=pos, color='r', label='Forecast start')
ax.legend()
```
